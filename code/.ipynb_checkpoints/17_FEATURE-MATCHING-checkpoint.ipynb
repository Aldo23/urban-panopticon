# FEATURE MATCHING

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display (img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
reeses = cv2.imread('data/reeses_puffs.png',0)#read as grayscale
display(reeses)
```

```python
cereals = cv2.imread('data/many_cereals.jpg',0)
display(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 =orb.detectAndCompute(reeses,None)
kp2,des2 =orb.detectAndCompute(cereals,None)
```

```python
# now we apply the brute force matching
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)
```

```python
matches = bf.match(des1,des2)
```

```python
#matches
```

```python
# check what is it this distance
single_match = matches[0]
```

```python
single_match.distance
```

```python
#like this we sort the matches according to their value
matches = sorted(matches, key=lambda x:x.distance)
```

```python
len(matches)
```

```python
reeses_matches = cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:30],None,flags=2)
```

```python
display(reeses_matches)
```
