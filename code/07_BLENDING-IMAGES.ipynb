```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('data/man.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('data/mask.jpg',cv2.IMREAD_UNCHANGED)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
# BLENDING IMAGES OF THE SAME SIZE 
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.5,gamma=0)
plt.imshow(blended)
```
