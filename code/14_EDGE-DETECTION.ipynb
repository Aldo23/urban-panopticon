```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('data/puppy-face.jpg')
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
plt.imshow(edges)
```

```python
#we will use now a formula to calculate the threshold
med_val = np.median(img)
med_val
```

```python
# lower threshold to either 0 or 70% of the median value
lower = int(max(0,0.7*med_val))

# upper threshold set to either 130% of the media or the max 255
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper+100)
plt.imshow(edges)
```

```python
# we will now try by blurring the image
blurred_img = cv2.blur(img,ksize=(5,5))

edges = cv2.Canny(image = blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
