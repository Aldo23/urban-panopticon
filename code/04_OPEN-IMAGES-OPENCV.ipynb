```python
new_img=cv2.resize(fix_img,(500,200))
plt.imshow(new_img)
```

```python
# NOW LETS RESIZE ACCORDING TO A SPECIFIC RATIO 
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(fix_img, (0,0),fix_img,w_ratio,h_ratio) # (0,0) IS A TUPLE
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,-1)
plt.imshow(new_img)
```
