```python
import sys
import folium
from folium.plugins import HeatMap
from folium.plugins import HeatMapWithTime
import numpy as np
import pandas as pd
import branca
from pyproj import Proj, transform
```

```python
#Import data
df_data = pd.read_csv('https://raw.githubusercontent.com/neriiacopo/test/master/test_points.csv')

df_data
```

```python
df_data3857 = pd.read_csv("C:/Users/Usuario/Desktop/IAAC/urban-panopticon/code/data/data.csv", header=None, names=("X","Y"))

# Convert CRS to 4326 for map projection
inProj = Proj('epsg:3857')
outProj = Proj('epsg:4326')

df_data = df_data3857
for i in range(0,len(df_data3857)):
   p = transform(inProj,outProj,df_data3857.iloc[i]["X"],df_data3857.iloc[i]["Y"])
   df_data.iloc[i] = p

df_data.columns = "lat","lng"
df_data
```

```python
# Set basemap
m = folium.Map(
    location=[41.388, 2.187],
    zoom_start=17,
    tiles='cartodbpositron')

# Circle Marker
for i in range(0,len(df_data)):
   folium.Circle(
       location=[df_data.iloc[i]['lat'], df_data.iloc[i]['lng']],
       radius=1,
       color='orange'
   ).add_to(m)

# Display the map
m
```

```python
# Set basemap
m = folium.Map(
    location=[41.388, 2.187],
    zoom_start=17,
    tiles='cartodbpositron')

# Popup Marker
for i in range(0,len(df_data)):
   folium.Marker(
       location=[df_data.iloc[i]['lat'], df_data.iloc[i]['lng']],
       popup=df_data.iloc[i]['type']
   ).add_to(m)

# Display the map
m
```

```python
# Set basemap
m = folium.Map(
    location=[41.388, 2.187],
    zoom_start=18,
    tiles='cartodbpositron')

# Popup Marker color-variable
for i in range(0,len(df_data)):
    if df_data.iloc[i]['type']=='ped':
        color='lightblue'
    elif df_data.iloc[i]['type']=='bike':
        color='blue'
    elif df_data.iloc[i]['type']=='car':
        color='darkblue'
        
    folium.Marker(
       location=[df_data.iloc[i]['lat'], df_data.iloc[i]['lng']],
       icon=folium.Icon(color=color)       
    ).add_to(m)
    
# Display the map
m
```

```python
# Set basemap
m = folium.Map(
    location=[41.388, 2.187],
    zoom_start=18,
    tiles='cartodbpositron')

# Popup Marker with fontawesome icons
# check https://fontawesome.com/v4.7.0/ for the full list _ remember to use V4.0 !
for i in range(0,len(df_data)):
    if df_data.iloc[i]['type']=='ped':
        icon='male',
        color='lightblue'
    elif df_data.iloc[i]['type']=='bike':
        icon='bicycle'
        color='blue'
    elif df_data.iloc[i]['type']=='car':
        icon='car',
        color='darkblue'
        
    folium.Marker(
       location=[df_data.iloc[i]['lat'], df_data.iloc[i]['lng']],
       icon=folium.Icon(
           prefix='fa',
           icon=icon,
           color=color)       
    ).add_to(m)

# Display the map    
m
```
